```python
import urllib.request
import os
import zipfile
import random
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers
from tensorflow.keras import Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.optimizers import RMSprop
from shutil import copyfile
import numpy as np
```

```python
import os
import random
from shutil import copyfile
import shutil  # Import the shutil module

def split_data(source, training, testing, split_size):
    files = os.listdir(source)
    random.shuffle(files)
    split_index = int(len(files) * split_size)
    training_files = files[:split_index]
    testing_files = files[split_index:]
    
    # Create directories for training and testing if they don't exist
    os.makedirs(training, exist_ok=True)
    os.makedirs(testing, exist_ok=True)
    
    # Copy files to training directory
    for file in training_files:
        src = os.path.join(source, file)
        dest = os.path.join(training, file)
        shutil.copyfile(src, dest)
    
    # Copy files to testing directory
    for file in testing_files:
        src = os.path.join(source, file)
        dest = os.path.join(testing, file)
        shutil.copyfile(src, dest)

# Define paths for the raw image data
person1_source = './bishwanath'
person2_source = './shivam'
person3_source = './gobinda'
train_dir = './train_data'
test_dir = './test_data'

# Create directories for training and testing data if they don't exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Split the data for each person into training and testing sets
split_size = 0.8  # 80% training, 20% testing
split_data(person1_source, os.path.join(train_dir, 'bishwanath'), os.path.join(test_dir, 'bishwanath'), split_size)
split_data(person2_source, os.path.join(train_dir, 'shivam'), os.path.join(test_dir, 'shivam'), split_size)
split_data(person3_source, os.path.join(train_dir, 'gobinda'), os.path.join(test_dir, 'gobinda'), split_size)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

TRAINING_DIR = './train_data'
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2  # If you want to split data into training and validation sets
)

train_generator = train_datagen.flow_from_directory(
    TRAINING_DIR,
    target_size=(150, 150),
    batch_size=100,
    class_mode='binary',
    subset='training'  # If you're using validation_split, set this to 'training'
)

# Retrieve the total number of classes
num_classes = train_generator.num_classes

VALIDATION_DIR = "./test_data"

validation_datagen = ImageDataGenerator(rescale=1./255)
validation_generator = validation_datagen.flow_from_directory(
    VALIDATION_DIR,
    batch_size=100,
    class_mode='binary',
    target_size=(150, 150)
)
```

```python
weights_url = "https://storage.googleapis.com/mledu-datasets/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5"
weights_file = "inception_v3.h5"
urllib.request.urlretrieve(weights_url, weights_file)

# Instantiate the model
pre_trained_model = InceptionV3(input_shape=(150, 150, 3),
                                include_top=False,
                                weights=None)

# load pre-trained weights
pre_trained_model.load_weights(weights_file)

# freeze the layers
for layer in pre_trained_model.layers:
    layer.trainable = False

# pre_trained_model.summary()

last_layer = pre_trained_model.get_layer('mixed7')

last_output = last_layer.output
```

```python
# Flatten the output layer to 1 dimension
x = layers.Flatten()(last_output)
# Add a fully connected layer with 1,024 hidden units and ReLU activation
x = layers.Dense(1024, activation='relu')(x)
X = layers.Dense(1024, activation='relu')(x)
# Add a final sigmoid layer for classification
x = layers.Dense(3, activation='softmax')(x)

model = Model(pre_trained_model.input, x)
```

```python
model.summary()
```

```python
from keras.optimizers import RMSprop

# Compile your model with appropriate loss and metrics for multi-class classification
model.compile(optimizer=RMSprop(learning_rate=0.0001),
              loss='sparse_categorical_crossentropy',
              metrics=['acc'])

# Train the model (adjust the number of epochs from 1 to improve performance)
history = model.fit(
    train_generator,
    epochs=150,
    validation_data=validation_generator,
    batch_size=10,
    verbose=1
)
model.save("bishwanathjanaModel.h5")
```

```python
import matplotlib.pyplot as plt

# Plot training & validation loss values
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper left')


plt.show()
```
